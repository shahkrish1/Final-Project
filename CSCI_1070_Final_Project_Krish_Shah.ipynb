```python
# Data Cleaning

# Remove unused columns
air_quality_data.drop(columns=['Unnamed: 15', 'Unnamed: 16'], inplace=True)

# Replace commas with dots and convert object columns to numeric where necessary
numeric_columns = ['CO(GT)', 'C6H6(GT)', 'T', 'RH', 'AH']
for col in numeric_columns:
    air_quality_data[col] = air_quality_data[col].str.replace(',', '.').astype(float)

# Confirm data cleaning
air_quality_data.info(), air_quality_data.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


# Select only numeric columns for correlation analysis
numeric_data = air_quality_data.select_dtypes(include=['number'])

# Compute the correlation matrix for numeric data only
correlation_matrix = numeric_data.corr()

# Create a heatmap to visualize correlations
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix of Air Quality and Environmental Metrics")
plt.show()
```

```python
import scipy.stats as stats
import statsmodels.api as sm
import numpy as np

# Define dependent variables (pollutants) and independent variables (environmental factors)
pollutants = ['CO(GT)', 'NOx(GT)', 'C6H6(GT)']
environmental_factors = ['T', 'RH', 'AH']

# Initialize dictionary to store correlation results
correlation_results = {}

# Perform Correlation Analysis
for pollutant in pollutants:
    for factor in environmental_factors:
        # Pearson correlation and p-value
        # Get the data for the current pollutant and factor, removing NaN and inf values
        valid_data = air_quality_data[[pollutant, factor]].replace([np.inf, -np.inf], np.nan).dropna()

        if len(valid_data) > 1:  # Check if there are at least 2 data points for correlation
            corr, p_value = stats.pearsonr(valid_data[pollutant], valid_data[factor])
            correlation_results[(pollutant, factor)] = (corr, p_value)
        else:
            print(f"Warning: Not enough valid data to calculate correlation for {pollutant} and {factor}.")
            correlation_results[(pollutant, factor)] = (np.nan, np.nan)
# Display correlation results
correlation_results_cleaned = {key: value for key, value in correlation_results.items() if not pd.isnull(value[0])}
correlation_results_cleaned

```

```python
import pandas as pd
import scipy.stats as stats
import statsmodels.api as sm
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Regression Analysis
for pollutant in pollutants:
    X = air_quality_data[environmental_factors]  # Independent variables
    y = air_quality_data[pollutant]  # Dependent variable

    # Remove rows with NaN or inf in either X or y
    # This creates a mask of rows to keep
    mask = ~(X.isin([np.inf, -np.inf]).any(axis=1) | y.isin([np.inf, -np.inf]) | X.isnull().any(axis=1) | y.isnull())

    X = X[mask]
    y = y[mask]

    # Add a constant for the intercept
    X = sm.add_constant(X)

    # Fit the regression model
    model = sm.OLS(y, X).fit()
    print(f"\nRegression Results for {pollutant}")
    print(model.summary())
```
